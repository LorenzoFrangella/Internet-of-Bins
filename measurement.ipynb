```python
import csv
import numpy as np
import matplotlib.pyplot as plt

# Read data from CSV file
data = []
with open('/Users/lorenzo/Desktop/data_cassonetto_ok.csv', 'r') as file:
    reader = csv.reader(file)
    for row in reader:
        data.append(row)

# Convert data to numpy arrays
time = np.array([float(row[0]) for row in data])
time = time-time[0]
values = np.array([float(row[1]) for row in data])

window_size = 5
moving_average = np.convolve(values, np.ones(window_size)/window_size, mode='valid')
time_ma = time[window_size-1:]

print("average value of energy consumtion is: ", values.mean())

plt.plot(time, values)
plt.xlabel('Time [s]')
plt.ylabel('Power consumption [mW]')
plt.title('Power consumption in activation phase')
plt.show()


```

```python
window_size = 20
moving_average = np.convolve(values, np.ones(window_size)/window_size, mode='valid')
time_ma = time[window_size-1:]

plt.plot(time_ma, moving_average)
plt.xlabel('Time [s]')
plt.ylabel('Power consumption [mW]')
plt.title('Power consumption in activation phase moving average')
plt.show()
```
